```python
# permite a utilização de caminhos relativos ao projeto, mesmo no diretório de notebooks

from knapsax.utils import setrootdir

setrootdir("knapsax")
```

```python
# Tou com dificuldades para encontrar esse caminho relativo. 
# EM todo caso fiz uma gambiarra navegando ate a apasta knapsax onde tem o ficheior ag.py
from knapsax.ga import AGenetico
from knapsax.optimization import Knapsack
from knapsax.stats import Stats
```

```python
knapsack = Knapsack(instance_file="data/knapsack-instance.txt")
knapsack
```

```python
ga_stats = Stats(
    n_runs=20,
    algorithm_class=AGenetico,
    knapsack=knapsack,
    nr_geracoes=100,
    taxa_mutacao=.05
)

ga_stats.run()
```

```python
df_ga = ga_stats.frame()
df_ga.to_csv("results/ga_stats.csv", index=False)
df_ga
```

```python
ga_stats.plot_convergence(
    title="Convergência do Algoritmo Genético",
    savefig="./results/ga_convergence.pdf"
)
```
