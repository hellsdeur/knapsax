```python
# Tou com dificuldades para encontrar esse caminho relativo. 
# EM todo caso fiz uma gambiarra navegando ate a apasta knapsax onde tem o ficheior ag.py
from knapsax.ga import AGenetico
from knapsax.optimization import Knapsack
from knapsax.stats import Stats
```

```python
knapsack = Knapsack(instance_file="data/knapsack-instance.txt")
knapsack
```

```python
ga = AGenetico(knapsack, nr_geracoes=55)
```

```python
import numpy as np
resultados = []
for i in range(20):
    ga = AGenetico(knapsack, nr_geracoes=55)
    _, best_value, _ = ga.run()
    resultados.append(best_value)

resultados = np.array(resultados)
print(f"\nMédia: {np.mean(resultados):.2f}")
print(f"Desvio padrão: {np.std(resultados):.2f}")
print(f"Melhor valor: {np.max(resultados)}")
print(f"Pior valor: {np.min(resultados)}")
```

```python
ga_stats = Stats(
    n_runs=20,
    algorithm_instance=ga
)

ga_stats.run()
```

```python
df_ga = ga_stats.frame()
df_ga
```

```python
ga_stats.plot_convergence(
    title="Convergência do Algoritmo Genético",
    savefig="./results/ga_convergence.png"
)
```
